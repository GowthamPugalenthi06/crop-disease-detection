### Step 2: Create `src/rename_labels.py` (using manual `LABEL_MAP`)

````python
import os

src_dir = '/content/crop-disease-detection/src'
rename_script_content = '''
import os
import shutil
from label_map import LABEL_MAP

def rename_dataset(input_path, output_path):
    os.makedirs(output_path, exist_ok=True)

    for class_name in os.listdir(input_path):
        class_path = os.path.join(input_path, class_name)

        if not os.path.isdir(class_path):
            continue

        if class_name not in LABEL_MAP:
            print(f"Skipping: {class_name}")
            continue

        new_class = LABEL_MAP[class_name]
        new_class_path = os.path.join(output_path, new_class)

        os.makedirs(new_class_path, exist_ok=True)

        for img in os.listdir(class_path):
            src = os.path.join(class_path, img)
            dst = os.path.join(new_class_path, img)

            # Using shutil.copy2 to preserve metadata if desired
            shutil.copy2(src, dst)

    print(f"Done processing: {input_path}")
'''

with open(os.path.join(src_dir, 'rename_labels.py'), 'w') as f:
    f.write(rename_script_content)

print("Created src/rename_labels.py (manual mapping version)")
````

### Step 4: Create `src/auto_label.py` for automated label normalization

````python
import os

src_dir = '/content/crop-disease-detection/src'
auto_label_content = '''
import re

def normalize_label(label):
    label = label.lower()

    # remove special characters
    label = re.sub(r'[^a-z\s]', '', label)

    words = label.split()

    # remove useless words
    remove_words = ["leaf", "plant", "disease"]
    words = [w for w in words if w not in remove_words]

    # standard replacements
    replacements = {
        "septoria": "Septoria",
        "blight": "Blight",
        "early": "Early",
        "late": "Late",
        "rust": "Rust",
        "scab": "Scab",
        "spot": "Spot",
        "bacterial": "Bacterial",
        "mosaic": "Mosaic",
        "yellow": "Yellow",
        "virus": "Virus",
        "black": "Black",
        "rot": "Rot"
    }

    final_words = []
    for w in words:
        if w in replacements:
            final_words.append(replacements[w])
        else:
            final_words.append(w.capitalize())

    return "_".join(final_words)
'''

with open(os.path.join(src_dir, 'auto_label.py'), 'w') as f:
    f.write(auto_label_content)

print("Created src/auto_label.py")
````

### Step 5: Update `src/rename_labels.py` (to use auto-normalization)

````python
import os

src_dir = '/content/crop-disease-detection/src'
updated_rename_script_content = '''
import os
import shutil
from auto_label import normalize_label

def rename_dataset(input_path, output_path):
    os.makedirs(output_path, exist_ok=True)

    for class_name in os.listdir(input_path):
        class_path = os.path.join(input_path, class_name)

        if not os.path.isdir(class_path):
            continue

        new_class = normalize_label(class_name)

        new_class_path = os.path.join(output_path, new_class)
        os.makedirs(new_class_path, exist_ok=True)

        for item in os.listdir(class_path):
            item_path = os.path.join(class_path, item)
            if os.path.isdir(item_path): # Skip directories
                continue
            src = item_path
            dst = os.path.join(new_class_path, item)

            shutil.copy2(src, dst)

        print(f"{class_name} \u2192 {new_class}") # Using unicode arrow
    print(f"Done processing: {input_path}")
'''

with open(os.path.join(src_dir, 'rename_labels.py'), 'w') as f:
    f.write(updated_rename_script_content)

print("Updated src/rename_labels.py (auto-normalization version)")
````
